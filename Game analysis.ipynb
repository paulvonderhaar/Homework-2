```python
#Import required library
import pandas as pd
```

```python
#My data lives here, if it needs to be modified let me know, the instructions were somewhat unclear
file="data.csv"
```

```python
#Define our first data frame, and convert the columns into lower case uniform strings
df=pd.read_csv(file)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df.head()
```

```python
#Get the total number of transactions
length=df.sn.count()
results=pd.DataFrame({"Total number of transactions":length},index=[0])
results
```

```python
#Item Analysis

#Total count of items
itemCounts=pd.DataFrame()

a=(len(df['item_id'].unique()))
b=(df['price'].mean())
b=round(b,2)
results=pd.DataFrame({'Number of Unique Items':a,'Average Price':b}, index=[0])
results
```

```python
#Getting the shape of the data frame, and the count of gender demographics, removing any player
#who has made more than one purchase
df_unique=df.drop_duplicates(subset="sn")


#Getting the new gender count dataframe
gender_demographics=pd.DataFrame()
gender_demographics['Total Count']=df_unique['gender'].value_counts()
gender_demographics['Percent of players']=round(100*gender_demographics['Total Count']/(df_unique.gender.count()),2)
gender_demographics
```

```python
#Doing Sales analysis by gender, relying on group by
gender_sales=pd.DataFrame()
gender_sales['Average Price']=df.groupby("gender")['price'].mean()
gender_sales['Total Revenue']=df.groupby("gender")['price'].sum()
gender_sales['Total number of Transactions']=df.groupby('gender')['price'].count()
#To find the average number of transactions we took the total number of transactions for each gender and divided it by the
#number of each gender found earlier
gender_sales['Average number of transactions per person']=round(gender_sales['Total number of Transactions']/gender_demographics['Total Count'],2)
gender_sales
```

```python
#Now we want to bin the data based on age, we will use the same df_unique table to avoid duplication of individuals. 
bin_cuts=[]
#Defining our bins
for i in range(0,52,4):
    bin_cuts.append(i)

```

```python
#Similair as beofre, we need to make sure no player is couted twice, so we cut df_unique into bins, and then use a value_counts()
#Function to create the age data table, we stopped the fore loop earlier when there were no more players in the bin
df_unique['bin']=pd.cut(df_unique.age,bin_cuts)
reportTable=pd.DataFrame()
reportTable['Age Count']=df_unique.bin.value_counts()
reportTable=reportTable.sort_index()
reportTable
```

```python
#Now we do the smae analysis as we did above, except with age
df['bin']=pd.cut(df.age,bin_cuts)
age_purchasing=pd.DataFrame()
age_purchasing['purchase count']=df['bin'].value_counts()
age_purchasing['average purchase']=round(df.groupby('bin')['price'].mean(),2)
age_purchasing['total revenue']=round(df.groupby('bin')['price'].sum(),2)
age_purchasing=age_purchasing.sort_index()
age_purchasing['Average number of purchases']=round(age_purchasing['purchase count']/reportTable['Age Count'],2)
age_purchasing.head()
```

```python
#Individual top sellers
top_sellers=pd.DataFrame()
top_sellers["number of purchases"]=df['sn'].value_counts()
top_sellers["Average value of purchase"]=round(df.groupby("sn")["price"].mean(),2)
top_sellers["Total purchase value"]=round(df.groupby("sn")["price"].sum(),2)
top_sellers=top_sellers.sort_values(by="Total purchase value", ascending=False)
top_sellers.head()
```

```python
#Item analysis was a little harder. First we took each unique item id, and then used a join command to join each item with its
#Item name. We dropped a collumn because we had an extra one, and then we did the analysis as above.
items_sold=pd.DataFrame()
items_sold["item_id"]=df["item_id"].unique()
items_sold=items_sold.join(df[['item_id',"item_name"]], on="item_id", lsuffix='_other')
items_sold=items_sold.drop('item_id_other',axis=1)
items_sold['Purchase Count']=df['item_id'].value_counts()
#I noticed in the data that the price is not uniform across the items, so I take the average price
#Assuming sales or something like that
items_sold['Average Item Price']=df.groupby('item_id')['price'].mean()
items_sold["Total profit from item"]=df.groupby('item_id')['price'].sum()
items_sold=items_sold.sort_values(by='Purchase Count', ascending=False)

items_sold.head()
```
